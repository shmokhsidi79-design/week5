#Creating dataset

```python

reading_lists = [
    # Fantasy
    ["harry_potter", "lotr", "hobbit", "game_of_thrones"],
    ["lotr", "hobbit", "wheel_of_time", "earthsea"],
    ["harry_potter", "percy_jackson", "hunger_games"],

    # Self-development
    ["atomic_habits", "deep_work", "the_power_of_habit", "mindset"],
    ["rich_dad_poor_dad", "the_psychology_of_money", "think_and_grow_rich"],
    ["zero_to_one", "the_lean_startup", "good_to_great"],

    # Classics
    ["pride_and_prejudice", "jane_eyre", "wuthering_heights"],
    ["to_kill_a_mockingbird", "1984", "animal_farm"],
    ["the_great_gatsby", "of_mice_and_men", "catcher_in_the_rye"],

    # Science
    ["sapiens", "homo_deus", "guns_germs_and_steel"],
    ["a_brief_history_of_time", "cosmos", "the_selfish_gene"],
]

print("Number of reading lists:", len(reading_lists))
print("Example list:", reading_lists[2])
```

```python
from gensim.models import Word2Vec
from gensim.models.phrases import Phrases, Phraser
```

```python
model = Word2Vec(
    sentences=reading_lists,
    vector_size=30,
    window=3,
    min_count=1,
    workers=2,
    sg=1
)
```

# books rec

```python
def recommend_books(book, top_n=4):
    if book not in model.wv:
        print(f"'{book}' not found. Try one of these:\n{list(model.wv.index_to_key)[:15]}")
        return

    print(f"Recommendations for '{book}':\n")
    for b, score in model.wv.most_similar(book, topn=top_n):
        print(f"- {b} (similarity: {score:.2f})")
```

```python
recommend_books("harry_potter")
recommend_books("atomic_habits")
recommend_books("1984")
recommend_books("sapiens")
```
