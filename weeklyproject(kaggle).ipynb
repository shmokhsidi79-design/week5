```python
import pandas as pd

pd=pd.read_csv("/content/zoo.csv")
pd
```

```python
sentences = (
    pd.groupby("class_type")["animal_name"]
    .apply(list)
    .tolist()
)

sentences
```

```python
!pip install gensim
```

```python
from gensim.models import Word2Vec

model = Word2Vec(
    sentences,
    vector_size=32,
    window=5,
    min_count=1,
    workers=4
)

model.wv.most_similar("lion")
```
